```python
# Importing libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import os
```

```python
matplotlib.__version__
```

```python
%matplotlib inline
```

```python
# Importing data

path = r'C:\Users\katro\World Happiness Report'
```

```python
df = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'whr.csv'))
```

```python
# Checking column names

df.columns
```

```python
# Creating a subset
sub = df[['Happiness Score', 'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Generosity']]
```

```python
sub
```

```python
# Creating a correlation heatmap with seaborn

f,ax = plt.subplots(figsize=(10,10))

corr = sns.heatmap(sub.corr(), annot = True, ax = ax)
```

Economy, health, and family have the strongest positive correlations to the happiness score. Generosity seems to have little impact on the happiness score.

```python
# Creating a scatterplot to compare Happiness Score and Economy (GDP per Capita)

splot_ec = sns.lmplot(x = 'Economy (GDP per Capita)', y = 'Happiness Score', data = df)
```

The Economy plot points are clustered closer to the Happiness Score which shows a stronger positive correlation.

```python
# Creating a scatterplot comparing Happiness Score and Health (Life Expectancy)

splot_h = sns.lmplot(x = 'Health (Life Expectancy)', y = 'Happiness Score', data = df)
```

The plot points for Health are showing a positive correlation, but the plot points arent closely clustered which shows it isn't very strong.

```python
# Creating a scatterplot comparing Happiness Score and Family

splot_fam = sns.lmplot(x = 'Family', y = 'Happiness Score', data = df)
```

While the plot points show a positive correlation, the link isn't very strong.

```python
# Dropping column 'Unnamed: 0'

df = df.drop(columns = ['Unnamed: 0'])
```

```python
df.columns
```

```python
# Creating pair plots

pairplt = sns.pairplot(df)
```

I want to keep exploring Economy, Health, and Family correlations.

```python
# Using a histogram to visualize the distribution of the variable

sns.histplot(df['Happiness Score'], bins = 20, kde = True)
```

```python
# Creating categories of the Happiness Score

df.loc[df['Happiness Score'] < 4, 'Happiness Range'] = 'Low Happiness'
```

```python
df.loc[(df['Happiness Score'] >= 4) & (df['Happiness Score'] < 6), 'Happiness Range'] = 'Medium Happiness'
```

```python
df.loc[df['Happiness Score'] >= 6, 'Happiness Range'] = 'High Happiness'
```

```python
df['Happiness Range'].value_counts(dropna = False)
```

```python
# Creating a categorical plot in seaborn using the happiness ranges created above

sns.set(style="ticks")
cplot = sns.catplot(x="Economy (GDP per Capita)", y="Happiness Score", hue="Happiness Range", data=df)
```

There are many countries with lower economy scores that have high happiness. There are also  some countries with higher economy scores that have low happiness.
